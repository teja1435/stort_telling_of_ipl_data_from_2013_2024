```python
#Create a story around year wise changes in the overall IPL
from matplotlib import pyplot as plt
df = pd.read_csv("C:/Users/dharm/Downloads/IPL2024_Data_Cleaned.csv")
df_seperated = df[['Year','Player_Name', 'Runs_Scored', 'Catches_Taken', 'Wickets_Taken']]
total_runs_per_year = df_seperated.groupby('Year')['Runs_Scored'].sum()
plt.figure(figsize=(12,6))
total_runs_per_year.plot(kind='bar', color='b')
plt.title('Total runs')
plt.xlabel('Year')
plt.ylabel('Total runs')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt
total_runs_per_year = df_seperated.groupby('Year')['Catches_Taken'].sum()
plt.figure(figsize=(12,6))
total_runs_per_year.plot(kind='bar', color='b')
plt.title('Catches_Taken')
plt.xlabel('Year')
plt.ylabel('Catches_Taken')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt
total_runs_per_year = df_seperated.groupby('Year')['Wickets_Taken'].sum()
plt.figure(figsize=(12,6))
total_runs_per_year.plot(kind='bar', color='b')
plt.title('Wickets_Taken')
plt.xlabel('Year')
plt.ylabel('Wickets_Taken')
plt.grid(True)
plt.tight_layout()
plt.show()
```
